# Remove the outliers

```python
Q1 = insurance['bmi'].quantile(0.25)  #Cal 25% percentile
Q3 = insurance['bmi'].quantile(0.75)  #Cal 75% percentile
print(Q1, Q3)
```

```python
IQR = Q3- Q1 #Cal Inter Quartile Range
IQR
```

```python
ub = Q3 + 1.5*IQR  # Upper Bound / Limit
lb = Q1 - 1.5*IQR  # Lower Bound / Limit
print(ub, lb)
```

```python
# Removing outliers by considering the data present pnly between lb and ub.
insurance = insurance[(insurance['bmi']>=lb) & (insurance['bmi']<=ub)]
```

```python
# Removing outliers for another column
Q1 = insurance.past_consultations.quantile(0.25)
Q3 = insurance.past_consultations.quantile(0.75)
IQR = Q3 - Q1
insurance = insurance[(insurance.past_consultations >= Q1 - 1.5*IQR) & (insurance.past_consultations <= Q3 + 1.5*IQR)]
```

```python
# Removing outliers for another column
Q1 = insurance.Hospital_expenditure.quantile(0.25)
Q3 = insurance.Hospital_expenditure.quantile(0.75)
IQR = Q3 - Q1
insurance = insurance[(insurance.Hospital_expenditure >= Q1 - 1.5*IQR) & (insurance.Hospital_expenditure <= Q3 + 1.5*IQR)]
```

```python
# Removing outliers for another column
Q1 = insurance.Anual_Salary.quantile(0.25)
Q3 = insurance.Anual_Salary.quantile(0.75)
IQR = Q3 - Q1
insurance = insurance[(insurance.Anual_Salary >= Q1 - 1.5*IQR) & (insurance.Anual_Salary <= Q3 + 1.5*IQR)]
```

VIF --> Variance Inflation factor  -->  to measure the multicollinearity we use the VIF and values should be <5

```python
# Picking columns apart object datatypes and target variable.

col_list = []
for col in insurance.columns:
    if((insurance[col].dtype!='object') & (col!='charges')):
        col_list.append(col)

col_list
```

```python
# Importing the VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = insurance[col_list] #numerical data
vif_data = pd.DataFrame() # Creating empty DataFrame
vif_data['feature'] = X.columns # adding features column
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] # adding vif column
print(vif_data)
```

```python
# 'num_of_steps' column having higher variance comapare to all, it means multicolinearity is present. So first delete that feature from the data

insurance = insurance.drop(['num_of_steps'], axis=1)
# or
# if vif_data['VIF'].max() > 5:
#   insurance = insurance.drop(vif_data['feature'][vif_data['VIF'].idxmax()], axis = 1)
```

# Model building

```python
# Converting to the list of values from features column
col = list(vif_data['feature'].unique())
col
```

```python
# From above only 5 columns are having <5 VIF. So, picks those for model building.
x = insurance.loc[:,col]
y = insurance.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

#train_test
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.8, random_state=999)
```

```python
from sklearn.linear_model import LinearRegression
l_model = LinearRegression()    # Object creation for LinearRegression
```

```python
# training the model
l_model.fit(x_train, y_train)
```

```python
# Prediction Values
y_pred = l_model.predict(x_test)
y_pred
```

```python
diff = pd.DataFrame()
diff['Actual'] = y_test
diff['Predicted'] = y_pred
diff
```

```python
diff['Error'] = diff['Actual'] - diff['Predicted']
diff.reset_index(drop=True, inplace=True)
diff
```

```python
# Checking error metrics

from sklearn.metrics import *
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'MAE: {mae:.2f}, MAPE: {mape:.2f}, MSE: {mse:.2f}, rmse: {rmse:.2f}')
```

```python
# Check the goodness of fit of the model

from sklearn.metrics import r2_score
r2score = r2_score(y_test, y_pred)
r2score
```
